**Performing Data Analysis**

Before we start doing any kind classification, we need to split the data into 2 segments, a training set and a testing set.
We also included a seed to ensure that our results are easily reproducible. 
For this project, we decided on setting the seed as 727. 

```r
set.seed(727)

heart_split <- initial_split(heart_data, prop = 0.75, strata = num)
heart_train <- training(heart_split)
heart_test <- testing(heart_split)
```

After we split the data, we must scale and center it. This is because K-Nearest Neighbor classification can overpredict in favour of one variable when the variables are not of comparable scale. 

Since we are predicting the severity of the heart disagnostic based on age, resting heartbeat, etc. , we included them as classifiers within  our recipe. 

```r
set.seed(727)
heart_recipe <- recipe(num ~ age  + thalach + chol + trestbps , data = heart_data) |>
    step_scale(all_predictors()) |>
    step_center(all_predictors())
```

We do not know exactly what number of neighors to use yet, so we test a range between 2 neighors and 21.

```r
kvalues <- tibble(neighbors = seq(from = 2, to = 21, by = 1))
```

Since we are using KNN classification, it is important to find the best number of neighbors for classification.
In order to do this, we will use 5-fold-cross-validation. 

The following code will be used to determine the optimal number of neighors used for this approximation.


```r
set.seed(727)

heart_vfold <- vfold_cv (heart_train, v = 5, strata = 4)

knn_tune <- nearest_neighbor(weight_func = "rectangular", neighbors = tune())|>
            set_engine("kknn")|>
            set_mode ("classification")

heart_wkfl <- workflow ()|>
            add_model(knn_tune)|>
            add_recipe(heart_recipe)|>
            tune_grid(resamples = heart_vfold, grid = kvalues)|>
            collect_metrics()

```

In order to see the accuracy each K value gives, we need to filter our workflow to only contain the accuracy. We can then plot it as a line graph to visually determine the best K value within our range.

```r
accuracy_data <- heart_wkfl |>
            filter(.metric == "accuracy")
```

```r
options(repr.plot.height = 5, repr.plot.width = 6)
# This top part is to make the graph fit the screen better

cross_val_plot <- accuracy_data |>
    ggplot(aes (x = neighbors, y = mean))+
    geom_point()+
    geom_line(aes(x = neighbors, y = mean))+
    xlab("Number of K's")+
    ylab ("Accuracy")+
ggtitle("Accuracy of predictions v.s. Nearest Neighbors")

cross_val_plot
```

Looking at this graph, we see that the best number of neighbors is 20.

Therefore, we will proceed with this project using 20 neighbors for classification.


```r
set.seed(727)


knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = 20) |>
       set_engine("kknn") |>
       set_mode("classification")

heart_fit <- workflow() |>
       add_recipe(heart_recipe) |>
       add_model(knn_spec) |>
       fit(data = heart_train)




```

Now we can predict! We will use the predict function to generate predictions for all of the observations in our test set alongside the model we trained from our training set.

```r
set.seed(727)
heart_test_predictions <- predict(heart_fit , heart_test) |>
       bind_cols(heart_test)


heart_test_predictions
```

**Testing the Accuracy of our KNN Classifier**

To quantify whether or not this model can successfuly predict the severity of heart disease, we will use the metrics function. Understanding the quality of our model with the upcoming functions will help us answer our question of whether this dataset can be used to predict heart disease severity. 

```r
set.seed(727)
heart_metrics <- heart_test_predictions |>
metrics(truth = num, estimate = .pred_class)|>
  filter(.metric == "accuracy")
heart_metrics
```

In this case, we find the accuracy to be around %57. Finally, we will include a confusion matrix to determine the precision and recall.

```r
heart_matrix <- heart_test_predictions |> 
      conf_mat(truth = num , estimate = .pred_class)

heart_matrix
```

**Summary of Findings**:
- While the initial visualizations using histograms suggested there would be a considerable relationship between these variables and heart disease severity, both the accuracy and confusion matrix results indicate that this was not the case. An accuracy of 57% is higher than just guessing but it is not sufficient to use the model for real-world applications. Furthermore, the model was not able to reliably classify the vast majority of observations with a severity greater than 0 as reflected in the recall values. As such, we can say that the parameters provided in this report are not sufficient to create a model that can accurately predict heart disease.
- It appears that the model overpredicted for a severity of 0. It is possible this is because there was not enough of a relationship between higher severity levels and a change in our chosen variables. Thus, the greater number of observations with a severity of 0 could have influenced the model to over-predict for low severity. 

**Expectations vs. Findings**:
-    The findings of this model did not allign with our expectations. We assumed the standardization of the variables would highlight a relationship between severity and each variable. While the model was not completely innacurate, it did not meet the expectations of an effective predictor and does not support the idea that physiological measurements can predict heart disease. Unlike the previous literature, there was a much more limited set of variables and number of observations. It is possible that changing either of these factors would allign the findings of a futher study with the existing literature. 

**Potential Impact**:
-    While this report on its own was not able to create a model that predicted heart disease severity, it did create a model that reliably labeled everyone with a severity of 0 correctly. The nuance of physiological changes associated with heart disease may not be able to be captured entirely by a model and dataset as simple as ours. However, further refinement may be able to create a model that can differentiate a severity of 0 from higher values with a much higher accuracy. If an individual had access to common heart related metrics, like their blood pressure, or resting heartrate, they would be able to quickly determine whether there is reason to be worried about their heart health by using this theoretically improved model.
